```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepalLength', 'sepalWidth','petalLength','petalWidth', 'target'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
features = ['sepalLength', 'sepalWidth','petalLength','petalWidth']
x = df.loc[:, features].values
y = df.loc[:, ['target']].values
x = StandardScaler().fit_transform(x)
df_scaled = pd.DataFrame(data = x, columns =['sepalLength', 'sepalWidth','petalLength','petalWidth'])
df_scaled["target"] = df.target
df_scaled.head()
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
principalComponents = pca.fit_transform(df_scaled.drop(columns=["target"]))
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf["target"]=df_scaled.target

principalDf.head()

```

```python
for i in range(1,5):
    pca=PCA(n_components=i)
    principalComponents = pca.fit_transform(df_scaled.drop(columns=["target"]))
    print(pca.explained_variance_ratio_.cumsum())
```
